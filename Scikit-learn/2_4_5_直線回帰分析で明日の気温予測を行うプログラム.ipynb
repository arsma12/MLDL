```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#10
df = pd.read_csv('2_4_1_kion10y.csv', encoding="utf-8")

#(1)
train_year = (df[""] <= 2015)  #9
test_year = (df[""] >= 2016)   #1
interval = 6

#-------------------------------------------
#6(2)
def make_data(data):
    x = [] #
    y = [] #
    temps = list(data[""])
    
    for i in range(len(temps)):
        if i < interval: continue
        y.append(temps[i])
        xa = []
        for p in range(interval):
            d = i+p-interval
            xa.append(temps[d])
        x.append(xa)
    return (x,y)
#-------------------------------------------

train_x, train_y = make_data(df[train_year])
test_x, test_y = make_data(df[test_year])

#(3)
lr = LinearRegression(normalize=True)
lr.fit(train_x,train_y)                     # fit
pre_y = lr.predict(test_x)                  # predict

#(4)
plt.figure(figsize=(10,6),dpi=100)
plt.plot(test_y,c='r')     # 
plt.plot(pre_y,c='b')      # 
plt.savefig('2_4_5_tenki-kion-lr.png')
plt.show()


```
