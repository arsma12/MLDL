```python
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

#カスケードファイルを指定して検出器を作成
cascade_file = "haarcascade_frontalface_alt.xml"
cascade = cv2.CascadeClassifier(cascade_file)

#画像の読み込んでグレイスケールに変換する
img = cv2.imread("3_2_1_test3.jpg")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#画像を表示
#plt.imshow(cv2.cvtColor(img_gray,cv2.COLOR_BGR2RGB))
#plt.show()

#顔認識を実行
face_list = cascade.detectMultiScale(img_gray, minSize=(50,50))

#結果を確認
if len(face_list)==0:
    print("失敗")
    quit()

#認識した部分に印をつける
for (x,y,w,h) in face_list:
    print("顔の座標=", x,y,w,h)
    red = (0,0,255)
    cv2.rectangle(img, (x,y), (x+w,y+h), red, thickness=3)
    
#画像を出力
cv2.imwrite("3_2_1_face-detect.png",img)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show()
```
