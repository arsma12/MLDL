### 【手順】

### 0.Notebook内で利用する関数の定義

### 1.データ読込
- 欠損項目を確認する

### 2.データ前処理
- 事前に予測できない項目は、除外する。
- 目的変数を説明するのに有効そうな説明変数を見つける。
- 目的変数と説明変数の関係を確認するためのグラフを作成する。
- 質的変数が扱えない場合は、ダミー変数に置き換える。

### 3.モデル生成
- ロジスティック回帰モデルを利用する
- モデルパラメータ（重み）：$\mathbf{w} = (w_0, w_1, w_2, \dots, w_p)^{\mathrm{T}} $
- モデルへの入力：$\mathbf{x} = (1, x_1, x_2, \dots, x_p)^{\mathrm{T}} $
- 以上の2つが与えられたときのモデルの出力（確率値）：$\hat{Y}(\mathbf{x}; \mathbf{w}) =  f(\mathbf{w}^{\mathrm{T}} \mathbf{x})$
- $f$はシグモイド関数：$f(x) = \frac{1}{1 + \exp(-x)}$

### 4.モデル評価
- 混同行列を作成し、Accuracy、Recall、Precisionを求める

### 5.テストデータ予測


### ※データセット
- 「train.csv」（学習データ数：891行12列）
- 「test.csv」（テストデータ数：418行11列 ※Survived列がない）

| 列 | 項目 | 項目名 | 欠損有無 | 利用有無 |
| :--- | :--- | :--- | :---: | :---: |
|  1 | PassengerId |識別ID | - | - |
|  2 | Survived | 生存有無 | - | ○ |
|  3 | Pclass | チケットクラス | - | ○ |
|  4 | Name | 名前 | - | - |
|  5 | Sex | 性別 | - | ○ |
|  6 | Age | 年齢 | 欠損あり | ○ |
|  7 | SibSp | タイタニック号に乗る兄弟/配偶者の数 | - | ○ |
|  8 | Parch | タイタニック号に乗る親/子供の数 | - | ○ |
|  9 | Ticket | チケット | - | - |
| 10 | Fare | 旅客運賃 | 欠損あり | ○ |
| 11 | Cabin | 客室番号 | 欠損あり | - |
| 12 | Embarked| 乗船港 C =シェルブール、Q =クイーンズタウン、S =サザンプトン | 欠損あり | ○ |

```python
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import log_loss, accuracy_score
from sklearn.linear_model import SGDClassifier
import matplotlib.pyplot as plt
        
c = 1.0  # 正則化のハイパーパラメータ
test_size = 0.2 # テスト用データの割合

# 分類ラベル（list型）
targets = ['死亡', '生存']  # 0 : 死亡, 1 : 生存

# 不要な項目(Dictionary型)
drop_cols = {'train' : ['Survived', 'Name', 'Age', 'Ticket', 'Fare', 'Cabin'], 
             'test' : ['PassengerId', 'Name', 'Age', 'Ticket', 'Fare', 'Cabin']}
    
# データセットパス
train_file_path = './train.csv' # 学習用データ
test_file_path = './test.csv' # テスト用データ
res_file_path = './titanic_submission.csv' # 提出用データ

# -----------------------------------------------------------------------------
# 1.Titanicクラス 
# -----------------------------------------------------------------------------
class Titanic:
    
    # -------------------------------------------------------------------------
    # 1.コンストラクタ
    # 引数1：train_file_path:学習用データセットファイルパス
    # 引数2：test_file_path:テスト用データセットファイルパス
    # -------------------------------------------------------------------------
    def __init__(self, train_file_path, test_file_path):

        #print('\n ■ 1-1.初期処理')
        
        # -------------------------------------------------
        # 1-1.csv読込
        # 学習データ読込（0列目をindexとする）
        self.df_train = pd.read_csv(train_file_path, index_col=0, encoding='UTF-8')

        # テストデータ読込
        self.df_test = pd.read_csv(test_file_path, index_col=None, encoding='UTF-8')
        
        # -------------------------------------------------
        # 1-2.前処理
        # 学習用データのデータ前処理（欠損値・外れ値・One-Hot-Encoding・標準化）
        self.df_train_dummies = self.preprocess(self.df_train)
        
        # テスト用データのデータ前処理（欠損値・外れ値・One-Hot-Encoding・標準化）
        self.df_test_dummies = self.preprocess(self.df_test)
        
        # -------------------------------------------------
        # 1-3.ロジスティック回帰モデル初期化
        self.clf = linear_model.LogisticRegression(C=c, solver='liblinear', random_state=0)
        
        # 1-3.ロジスティック回帰モデル初期化
        #self.clf = SGDClassifier(loss='log',           # ロジスティック回帰を指定
        #            penalty='l2',                      # 正則化を入れる場合はL1､L2
        #            max_iter=10000,                    # 繰り返し計算時の強制的終了値、誤差が収束すれば利用しない。
        #            fit_intercept=False,                # 切片を有効にするかどうか
        #            random_state=1234,                 # 乱数のシード
        #            tol=1e-3                           # 収束する誤差の指定
        #           )
        
        # 回帰係数(重み)を初期化
        self.coef = []


    # -------------------------------------------------------------------------
    # 1-2.データ前処理（欠損値・外れ値・One-Hot-Encoding・標準化）
    # -------------------------------------------------------------------------
    def preprocess(self, df):

        #print('\n ■ 1-2.前処理')
        
        # 欠損値の処理
        df['Age'] = df['Age'].fillna(df['Age'].median())                  # 中央値
        df['Cabin'] = df['Cabin'].fillna(df['Cabin'].mode())              # 最頻値
        df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode())     # 最頻値
        df['Fare'] = df['Fare'].fillna(df['Fare'].median())               # 中央値
        
        # 外れ値の処理
        # Parch値 9 は学習用データにない為、学習用データの最大値 6 に置換
        df['Parch'] = np.where(df['Parch']==9, 6, df['Parch'])
        
        # ダミー変数
        df_dummies = pd.get_dummies(df, columns=['Sex', 'Pclass', 'SibSp', 'Parch', 'Embarked'])

        # 標準化
        scaler = StandardScaler()        
        df_dummies['Age_scale'] = scaler.fit_transform(df_dummies.loc[:, ['Age']])
        df_dummies['Fare_scale'] = scaler.fit_transform(df_dummies.loc[:, ['Fare']])
        
        return df_dummies
        
    # -------------------------------------------------------------------------
    # 2-1.データを確認
    # -------------------------------------------------------------------------
    def check_data(self):

        print('\n ■ 2-1.データ確認')
        
        # 欠損値確認
        #print("\n▼ 学習データの欠損項目を確認")
        #print(self.df_train.isnull().sum())
        
        #print("\n▼ テストデータの欠損項目を確認")
        #print(self.df_test.isnull().sum())
        
        # 基本統計量
        #print('\n▼ 2-1-1.学習データの基本統計量')
        #print(self.df_train.describe(), '\n')
        
        #print('\n▼ 2-1-2.テストデータの基本統計量')
        #print(self.df_test.describe(), '\n')
    
        #print('\n▼ 2-1-3.学習データ(DUMMY)の基本統計量')
        #print(self.df_train_dummies.describe(), '\n')
        
        #print('\n▼ 2-1-4.テストデータ(DUMMY)の基本統計量')
        #print(self.df_test_dummies.describe(), '\n')

        # カラム情報
        #print('\n▼ 2-1-5.学習データ(DUMMY)のカラム情報')
        #print(self.df_train_dummies.info(), '\n')
        
        #print('\n▼ 2-1-6.テストデータ(DUMMY)のカラム情報')
        #print(self.df_test_dummies.info(), '\n')

        
    # -------------------------------------------------------------------------
    # 2-2.目的変数・説明変数を作成
    # -------------------------------------------------------------------------
    def split_data(self, flag):
        
        #print('\n ■ 2-2.目的変数・説明変数を作成:' + flag)

        if flag == 'train':
            # ---------------------------------------------
            # 学習用データの場合
            # ---------------------------------------------
            # 説明変数（DictionaryからtrainのItemを取得）
            X = self.df_train_dummies.drop(drop_cols[flag], axis=1)
            
            # 目的変数
            y = self.df_train_dummies['Survived'].values
            
            print('\n▼2-2-1.学習データ説明変数')
            print(X.columns)
            #print('\n▼2-2-2.学習データ目的変数')
            #print(y)
            
            return X, y
        else:
            # ---------------------------------------------
            # テストデータの場合
            # ---------------------------------------------
            # 説明変数
            X = self.df_test_dummies.drop(drop_cols[flag], axis=1)
            
            print('\n▼2-2-1.テストデータ説明変数')
            print(X.columns)
            
            return X
    
    
    # -------------------------------------------------------------------------
    # 2-3.予測モデルを作成
    # -------------------------------------------------------------------------
    def model(self, X, y):
        
        # 学習
        #print('\n ■ 2-3.学習実行')
        self.clf.fit(X, y)
        
        # 回帰係数を抽出
        self.coef = self.clf.coef_
        
        #print('\n ■ 回帰係数表示')
        #print(self.coef)
        
    # -------------------------------------------------------------------------
    # 2-4.説明変数の回帰係数(重み)を表示
    # columns:列リスト
    # -------------------------------------------------------------------------
    def show_coef(self, columns):
        n_features = list(columns)
        fti = np.reshape(self.coef, -1).tolist() # 配列型へ変換
        
        #print('\n ■ 2-4-1.説明変数の回帰係数:')
        for i, feature in enumerate(columns):
            print('\t{0:10s} : {1:>.6f}'.format(feature, fti[i]))
        
        # グラフ化（説明変数の順番を逆に表示）  
        plt.figure(figsize=(10, 15))
        plt.barh(np.arange(len(n_features))[::-1], fti, align='center')
        plt.yticks(np.arange(len(n_features))[::-1], n_features)
        plt.title('Rotated Bar Chart')
        plt.xlabel('weight')
        plt.ylabel('X')
        plt.grid()
        plt.show()
        
        
    # -------------------------------------------------------------------------
    # 2-5.予測
    # X：説明変数
    # -------------------------------------------------------------------------
    def predict(self, X):
        
        # 予測値を計算
        y = self.clf.predict(X)
        
        return y
    
    
    # -------------------------------------------------------------------------
    # 2-6.モデルを検証
    # 検証データ：y_train2
    # 作成してモデルから求めた目的変数：y_train2_pred
    # -------------------------------------------------------------------------
    def verification(self, y_train2, y_train2_pred):

        print('\n ■ 2-6.モデルの検証')
        
        #print('\n ■ 検証データ:y_train2')
        #print(y_train2)
        
        #print('\n ■ 作成した予測モデルから求めた結果:y_train2_pred')
        #print(y_train2_pred)
        
        # 対数尤度を表示
        print('\n ■ 対数尤度 = {:.3f}'.format(- log_loss(y_train2, y_train2_pred)))
        
        # 正答率を表示
        print('\n ■ 正答率 = {:.3f}%'.format(100 * accuracy_score(y_train2, y_train2_pred)))

        # 混同行列
        confusion = confusion_matrix(y_train2, y_train2_pred)
        print('\n ■ 2-6-1.混同行列（検証用データ）')
        print(confusion, '\n')

        print('\n ■ 2-6-2.分類レポート（検証用データ）')
        print(classification_report(y_train2, y_train2_pred, 
                                    target_names=targets))
    
    
    # -------------------------------------------------------------------------
    # 2-7.提出用データを出力
    # predict:テストデータの予測結果
    # res_file_path:出力先ファイルパス
    # -------------------------------------------------------------------------
    def export_data(self, predict, res_file_path):
        
        # テストデータには、Survived列がないため、列を追加する
        print('\n▼ self.df_test')
        print(self.df_test.info())
        
        print('\n▼ predict')
        print(predict)
        
        # 予測結果とスコアをテストデータに追加
        self.df_test['Survived'] = predict
        
        # 提出用データをCSV出力する（idとSurvived列を出力）
        self.df_test.loc[:, ['PassengerId', 'Survived']].to_csv(
                res_file_path, index=False, header=True)
    
    
# -----------------------------------------------------------------------------
# メイン処理
# -----------------------------------------------------------------------------
def main(titanic):
    
    # 2-1.データを確認
    titanic.check_data()
    
    # 2-2.学習用データより目的変数・説明変数を作成
    (X, y) = titanic.split_data('train')
    
    # 2-3.tarinデータセットを学習用(8割)・検証用（2割）に分割　※ホールドｱｳﾄ法
    (X_train, X_train2, y_train, y_train2) = train_test_split(X, y,
                                           test_size=test_size, random_state=0)
    
    # 2-4.予測モデルを学習用データ（8割）を使って作成する
    titanic.model(X_train, y_train)
    
    # 2-5.説明変数の回帰係数を表示
    titanic.show_coef(X.columns)
    
    # 2-6.完成したモデルで検証データ（2割）を使って予測する
    y_train2_pred = titanic.predict(X_train2)
    
    # 2-7.モデルを検証
    titanic.verification(y_train2, y_train2_pred)
    
    # 2-8.テスト用データを予測
    y_test_pred = titanic.predict(titanic.split_data('test'))
        
    # 2-9.予測結果をエクスポート
    titanic.export_data(y_test_pred, res_file_path)

# -----------------------------------------------------------------------------
# 0.呼び出し元
# インポート時にプログラムが動かないようにするために、if __name__ == "__main__": と記載 
# import Titanicの場合、__name__は、'Titanic'が設定される。
# python Titanic.pyの場合、__name__は、'__main__'が設定される。
# -----------------------------------------------------------------------------
if __name__ == '__main__':
        
    #print("\n ■ 1.Titanicクラスを初期化（引数：学習ファイル、テストファイルのパス）")
    # 1.Titanicクラスを初期化（引数：学習ファイル、テストファイルのパス）
    titanic = Titanic(train_file_path, test_file_path) 
    
    #print("\n ■ 2.メイン処理：titanicインスタンスを引数にmain関数を実行する")
    # 2.メイン処理：titanicインスタンスを引数にmain関数を実行する
    main(titanic)
```
